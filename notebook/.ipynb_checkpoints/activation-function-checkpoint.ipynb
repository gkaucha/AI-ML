# Activation Function in Neural Networks
-----

## Preface

In an Artificial Neural Network(ANN), each neuron(a basic unit) in a layer is connected to some or all other neurons in the next layer. When the inputs are transmitted between neurons, the weights are applied to the inputs along with the bias.

A neuron(Y) can be represented as:

Y = $\sum$(weight * input) + bias

where
1. **weight** decides how much influence the input will have on the output. (the strenght of connection between neurons)
2. **bias** is a constant; an additional input into the next layer that will always have value of 1. (guarantees that even when all the inputs are zeros, there will still be an activation in the neuron)


## What is Activation Function?

As in the biological brain, which processes the input signals of different nature and intensity and decides whether the neuron should be fired("activated") or not. 

Similarly in the case of an artificial neuron, there has to be a mechanism(or function) that **normalizes the inputs and produces an output** which is then passed forward into the subsequent layer. That mechanism is referred to as Activation Function.

![activation function in neuron](./assets/neuron.jpg "Neuron and Activation Function")


## Why do neural networks need the activation function?

*Activation function adds non-linearity to the output which enables neural networks to solve non-linear problems.*

Let's suppose, we are not using any activation functions in the neural network. In that case, every neuron as shown in the above image will be performing a linear tranformation on the inputs using the weights and biases which is: 
z = wx + b

As the composition of two linear functions is a linear function itself, all layers in the network will behave in the same way. In other words, the neural network would be just a **linear regression** model.

## 3. Non-Linear Activation Function


### 3.1 Sigmoid / Logistic Function

This function takes any real value as input and outputs values in the range of 0 to 1. 

The larger the input (more positive), the closer the output value will be to 1.0, whereas the smaller the input (more negative), the closer the output will be to 0.0.

Mathematical Representation: 
$ f(x) = {1 \over 1 + e^-x} $


```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Activation Function
def sigmoid(x):
    return 1/(1+np.exp(-x))


# Creating Data for plot
x_data = np.linspace(-10,10)
y_data = sigmoid(x_data)

# Plotting Sigmoid
plt.plot(x_data, y_data)
plt.title('Sigmoid Activation Function')
plt.legend(['sigmoid'])
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```

- Becuase the range of output is between 0 and 1, can be used in model which predict probability as output
- Provides smooth gardient represented by S-shape of the function

#### Limitation : 
- Vanishing gradient problem (for values greater than 3 or less than -3)
- Output is not symmetric around zero causing neural networks training difficult.
 

## 3.2 Tanh function (Hyperbolic Tangent)

Similar to sigmoid but the difference is the output range of -1 to 1. The larger the input (more positive), the closer the output value will be to 1.0, whereas the smaller the input (more negative), the closer the output will be to -1.0.

Mathematical Representation: 
$$ f(x) = { e^x - e^-x \over e^x + e^-x } $$

- Is zero centered, so can map strongly negative, neutral or strongly positive
- Helps centering the data in hidden layers

#### Limitation:
- vanishing gardient issue, much steeper compared to sigmoid function


```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbolic Tangent (htan) Activation Function
def htan(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))


# Generating data for Graph
x_data = np.linspace(-6,6)
y_data = htan(x_data)


# Graph
plt.plot(x_data, y_data)
plt.title('tanh Activation Function')
plt.legend(['tanh'])
plt.grid()
plt.show()
```

## 3.3 ReLU Function

The rectified linear activation function (RELU) is a piecewise linear function that, if the input is positive say x, the output will be x. otherwise, it outputs zero.

Mathematical Representation:
$ f(x) = max(0,x) $

```python
## ReLU
def ReLU(x):
    data = [max(0,value) for value in x]
    return np.array(data, dtype=float)

# Generating data for Graph
x_data = np.linspace(-10,10)
y_data = ReLU(x_data)

# Graph Plotting
plt.plot(x_data, y_data)
plt.title('ReLU Activation Function')
plt.grid()
plt.show()
```

- allows backpropagation, computionally efficeint.
- accelearates the convergence of gardient descent towars the minimum loss function

#### Limitation:
- During backpropagation process, due to gradient value zero on negative inputs, dead neurons are created(weights and biases are not updated) which never get activated
- Model's ability to fit or train from data properly decreases.


## 3.4 Softmax Function

In scenarios where for multiple outputs, the sum of probability exceeds 1, sigmoid function cannot move forward as
the range of the function is limited to 0 and 1.

For calculating the probability of all the classes/output that exceeds the probability sum of 1, we need a **Softmax function** which is the combination of multiple sigmoids. This calculates the relative probabilities of each class.

Mathematical Representation:
   $$  softmax(z_i) = { exp(z_i) \over \sum_j exp(z_j) }    $$
   
Let's assume a  neural netork that classifies a given image, whether it is of cat, dog, wolf or none. It means we have four classes (4 neurons in the output layer). Let's suppose the output from the neurons is [1.8, 0.9, 0.68,0.19]

```python
def softmax(Z_data):
    exp_Z_data = np.exp(Z_data)
    #print("exp(Z_data) = ", exp_Z_data)
    
    sum_of_exp_Z_data = np.sum(exp_Z_data)
    #print("Sum of exponentials = ",sum_of_exp_Z_data)
    
    prob_dist = [exp_Zi/sum_of_exp_Z_data for exp_Zi in exp_Z_data]
    
    return np.array(prob_dist, dtype=float)
    
    
Z_data = [1.8,0.9,0.68,0.19]

#calculate probability of each class using softmax function

p_cat = softmax(Z_data)[0]
print("Probability of being cat:", p_cat)

print()
p_dog = softmax(Z_data)[1]
print("Probability of being a dog: ", p_dog)

print()
p_wolf = softmax(Z_data)[2]
print("Probability of being a wolf: ", p_wolf)

print()
p_none = softmax(Z_data)[3]
print("Probability of being none: ", p_none)
```

This function return 1 for the largest probability index and 0 for other indexes. 
In the above example, it would give full weight to index 0 and no weight to other indexes.
So the output would be the class corresponding to the index 0 which is cat.
