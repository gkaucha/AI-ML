```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-10,10)
y=x
plt.plot(x,y,c="red")
plt.title("Identity activation function y=f(x)")
plt.show()
```

```python
# binary activation function y=f(x) => y=1 if x>0, y=0 if x<0
x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    if(x[i]>0):
        y.append(1)
    else:
        y.append(0)

plt.plot(x,y,c="blue")
plt.title("binary activation function")
plt.show()
```

```python
#bipolar activation function  y=f(x) ==> y=1 if x>0, y=-1 if x<=0
x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    if(x[i]>0):
        y.append(1)
    else:
        y.append(-1)

plt.plot(x,y,c="red")
plt.grid()
plt.title("Bipolar Activation Function")
plt.hlines(y=0,xmin=-10, xmax=9)
plt.show()
```

```python
#sigmoid activation function y=1/(1+e^-x)
x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    y.append((1/(1+np.exp(-x[i]))))
    
plt.plot(x,y,c="blue")
plt.grid()
plt.title("Sigmoid Acivation Function")
plt.show()
```
